# Problem Statement 1.4 

```python
df['Month'] = pd.DatetimeIndex(df['Date']).month
df['Year'] = pd.DatetimeIndex(df['Date']).year

# Creating new columns Month and Year
```

```python
df.head()
```

```python
df['H+L+C'] = (df[['High Price', 'Low Price' , 'Close Price']].sum(axis=1))/3

# Price = (High Price + Low Price + Close Price) / 3 for the calculation of VWAP
```

```python
def vwap(df):
    q = df['Total Traded Quantity'].values
    p = df['H+L+C'].values
    return df.assign(vwap=(p * q).cumsum() / q.cumsum())

# ( VWAP = sum(price*volume)/sum(volume) ) 
# Volume = Total Traded Quantity
# Price = df['H+L+C'] = (High Price + Low Price + Close Price) / 3
```

```python
new_df = df.groupby(df['Month'], group_keys=False).apply(vwap)
```

```python
new_df.head()
```

```python
print('Shape of Given Data Frame: ', df.shape)
print('Shape of New Data Frame with VWAP:', new_df.shape)
```

# Problem Statement 1.5

```python
def avg(n):
    selected_date = pd.date_range(df['Date'].max() - pd.to_timedelta(n+1, unit='d'), df['Date'].max(), freq='D')
    df_selected = df[df["Date"].isin(selected_date)]
    
    df_avg = pd.DataFrame(df_selected.mean().to_dict(),index=[df_selected.index.values[-1]])

    return df_avg 

# Function to calculate the average price over the last N days
```

```python
avg(30)

# Finding avg of last 30 days
```

```python
def prof_loss(n):
    selected_date = pd.date_range(df['Date'].max() - pd.to_timedelta(n+1, unit='d'), df['Date'].max(), freq='D')
    df_selected = df[df["Date"].isin(selected_date)]
    
    df_selected['Prof/Loss'] = (df_selected['Close Price'] / df_selected['Close Price'].shift())*100 - 100


    return df_selected 

# Function to calculate the profit/loss percentage over the last N days.
```

```python
prof_loss(4)
```

```python
def avg_price(n):
    selected_date = pd.date_range(df['Date'].max() - pd.to_timedelta(n+1, unit='d'), df['Date'].max(), freq='D')
    df_selected = df[df["Date"].isin(selected_date)]
    
    average_price = (df_selected['Average Price'].tail(n).sum())/n

    return print('The average price for last n days is:',average_price)

# Function to calculate the average price over the last N days
```

```python
avg_price(5)
```

```python
def proloss(n):
    selected_date = pd.date_range(df['Date'].max() - pd.to_timedelta(n+1, unit='d'), df['Date'].max(), freq='D')
    df_selected = df[df["Date"].isin(selected_date)]
    
    df_selected['Prof/Loss'] = (df_selected['Close Price'] / df_selected['Close Price'].shift())*100 - 100
    overall_proloss = (df_selected['Prof/Loss'].tail(n).sum())/n


    return print('The profit/loss for last n days is:',overall_proloss)

# Function to calculate overall profit/loss percentage over the last N days.
```

```python
df['Prof/Loss'] = df['Close Price'].pct_change()*100   # 100 is multiplied since the next problem statement has the values accordingly.
df = df.fillna(0)
df['Day_Perc_Change'] = (df['Prof/Loss'] + df['Prof/Loss'].shift()) 
df = df.fillna(0)
df.head()
```

# Problem Statement 1.7 

```python
def trend(row):
    
    if row['Day_Perc_Change'] < -7 :
        return 'Bear drop'

    if row['Day_Perc_Change'] <= -3 and row['Day_Perc_Change'] > -7 :
        return 'Among top losers'

    if row['Day_Perc_Change'] <= -1 and row['Day_Perc_Change'] > -3 :
        return 'Negative'

    if row['Day_Perc_Change'] <= -0.5 and row['Day_Perc_Change'] > -1 :
        return 'Slight negative'

    if row['Day_Perc_Change'] <= 0.5 and row['Day_Perc_Change'] > -0.5 :
        return 'Slight or No change'

    if row['Day_Perc_Change'] >= 0.5 and row['Day_Perc_Change'] < 1 :
        return 'Slight positive'

    if row['Day_Perc_Change'] >= 1 and row['Day_Perc_Change'] < 3 :
        return 'Positive'

    if row['Day_Perc_Change'] >= 3 and row['Day_Perc_Change'] < 7 :
        return 'Among top gainers'

    if row['Day_Perc_Change'] > 7 :
        return 'Bull run'

# Function that returns required strings based on Day_Perc_Change column
```

```python
df.apply (lambda row: trend(row), axis=1)
df['Trend'] = df.apply (lambda row: trend(row), axis=1)

# Creating a new column Trends and importing strings developed based on Day_Perc_Change 
df.head()
```
